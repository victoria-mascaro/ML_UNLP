```python
# Librerías básicas para manipulación y análisis de datos
import pandas as pd
import numpy as np

# Librerías para visualización
import matplotlib.pyplot as plt
import seaborn as sns

# Librerías para preprocesamiento y evaluación
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import classification_report, confusion_matrix, mean_squared_error

# Modelos (regresión y clasificación)
from sklearn.linear_model import LinearRegression, ElasticNet, LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neural_network import MLPClassifier

# Otras utilidades
import warnings
warnings.filterwarnings("ignore")
```

```python
# path relativo de la carpeta "stores"
store_path = os.path.join("stores")
# Ruta relativa al archivo ZIP
zip_path = os.path.join("stores", "mlunlp-2024-ps-2.zip")
```

```python
# Directorio de destino (misma carpeta donde está el archivo ZIP)
extract_dir = os.path.dirname(zip_path)

# Verificar que el archivo ZIP existe
if not os.path.exists(zip_path):
    raise FileNotFoundError(f"El archivo {zip_path} no se encontró.")

# Descomprimir el archivo
with ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_dir)

print(f"Archivo descomprimido en: {extract_dir}")
```

```python
# Crear paths para cada archivo CSV
sample_submission_path = os.path.join(store_path, "sample_submission.csv")
train_hogares_path = os.path.join(store_path, "train_hogares.csv")
train_personas_path = os.path.join(store_path, "train_personas.csv")
test_hogares_path = os.path.join(store_path, "test_hogares.csv")
test_personas_path = os.path.join(store_path, "test_personas.csv")
```
