```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from google.colab import drive

drive.mount('/content/drive')
```

```python
# Define el path una sola vez
path = '/content/drive/MyDrive/UNLP_Machine_Learning/'
```

```python
# Número de página que deseas procesar
# Se puede convertir en un rango o una lista para iterar sobre varias páginas
number = 1

# Construir la URL para la página específica
url = f'https://ignaciomsarmiento.github.io/GEIH2018_sample/pages/geih_page_{number}.html'

try:
    # Hacer la solicitud
    response = requests.get(url)
    response.raise_for_status()  # Lanza una excepción para códigos de estado no exitosos

    print(f"Conexión exitosa a la página {number}")

    # Parsear el contenido HTML
    soup = BeautifulSoup(response.content, 'html.parser')

    # Extraer la tabla
    table = soup.find('table')

    if table:  # Verificar si se encontró la tabla
        # Extraer los encabezados
        headers = [header.text.strip() for header in table.find_all('th')]

        # Extraer las filas de datos de la tabla
        page_data = []
        for row in table.find_all('tr')[1:]:  # [1:] omite la fila de encabezado
            row_data = [cell.text.strip() for cell in row.find_all('td')]
            if row_data:
                page_data.append(row_data)

        # Crear un DataFrame para la página actual
        df = pd.DataFrame(page_data, columns=headers)

        # Guardar el DataFrame como CSV en la ruta especificada
        df.to_csv(path + f'page_{number}.csv', index=False)
        print(f"El archivo page_{number}.csv se ha guardado correctamente en {path}")

        # Mostrar el DataFrame (opcional)
        df.head(3)

    else:
        print(f"No se encontró ninguna tabla en la página {number}")

except requests.exceptions.RequestException as e:
    print(f"Error al conectar con la página {number}: {e}")
except Exception as e:
    print(f"Ocurrió un error: {e}")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import os
from google.colab import drive

drive.mount('/content/drive')

# Define el path una sola vez
path = '/content/drive/MyDrive/UNLP_Machine_Learning/'

# Crea la carpeta si no existe
os.makedirs(path, exist_ok=True)

# Nombre del archivo contenedor
combined_csv_filename = "combined_data.csv"
combined_csv_filepath = os.path.join(path, combined_csv_filename)

# Lista para almacenar todos los DataFrames
all_dataframes = []

for number in range(1, 11):  # Itera de la página 1 a la 10
    url = f'https://ignaciomsarmiento.github.io/GEIH2018_sample/pages/geih_page_{number}.html'

    try:
        response = requests.get(url)
        response.raise_for_status()
        print(f"Conexión exitosa a la página {number}")

        soup = BeautifulSoup(response.content, 'html.parser')
        table = soup.find('table')

        if table:
            headers = [header.text.strip() for header in table.find_all('th')]
            page_data = []
            for row in table.find_all('tr')[1:]:
                row_data = [cell.text.strip() for cell in row.find_all('td')]
                if row_data:
                    page_data.append(row_data)

            df = pd.DataFrame(page_data, columns=headers)
            all_dataframes.append(df) # Agregar el DataFrame a la lista

            # Guardar el archivo individual (opcional)
            # df.to_csv(path + f'page_{number}.csv', index=False)
            # print(f"Archivo page_{number}.csv guardado")

        else:
            print(f"No se encontró tabla en la página {number}")

    except requests.exceptions.RequestException as e:
        print(f"Error de conexión a la página {number}: {e}")
    except Exception as e:
        print(f"Error en la página {number}: {e}")

# Combinar todos los DataFrames
if all_dataframes:
    combined_df = pd.concat(all_dataframes, ignore_index=True)
    combined_df.to_csv(combined_csv_filepath, index=False, encoding='utf-8') # Especifica la codificación
    print(f"Archivo combinado '{combined_csv_filename}' guardado en '{path}'")
else:
    print("No se encontraron datos para combinar.")
```
